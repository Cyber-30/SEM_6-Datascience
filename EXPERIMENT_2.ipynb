The passport office claims that the passport application are processed withing 30 days of submitting the application form and all necessary documents. Write a program to conduct a hypothesis test at  significance level {alpha} = 0.05 to verify the claim made by the passport office

```python
import math
from scipy import stats # Import the stats module

def z_test(pop_mean, pop_std, sample):
  z_score = (sample.mean() - pop_mean)/(pop_std/math.sqrt(len(sample)))
  return z_score, stats.norm.cdf(z_score)

z_test(30, 12.5, passport_data)
```
